```python
df = df.drop('eventType',axis=1)
df = df.drop('halfInning',axis=1)
df = df.drop('batSide',axis=1)
df = df.drop('pitchType',axis=1)
df = df.drop('pitchHand',axis=1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df.drop('class_encoded', axis=1), df['class_encoded'], test_size=0.2, random_state=42)
```

```python
print(Y_train.value_counts())
```

```python
clf = DecisionTreeClassifier(max_depth=5, random_state=42)
clf.fit(X_train, Y_train)
```

```python
# Evaluate the model
accuracy = clf.score(X_test, Y_test)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
